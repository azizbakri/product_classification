# Start building the model

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import Flatten
from keras.layers.embeddings import Embedding
from tensorflow.keras.callbacks import EarlyStopping
from keras.layers import Conv1D, GlobalMaxPooling1D, MaxPooling1D
```

```python
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
```

# ANN with training the embedding layer

```python
classifier2 = Sequential()
e2 = Embedding(vocab_num+1, 100, weights=[embedding_matrix], input_length=max_len, trainable=True)
classifier2.add(e2)
classifier2.add(Dropout(0.2))
classifier2.add(Flatten())
classifier2.add(Dense(1024, activation='relu'))
classifier2.add(Dropout(0.5))
classifier2.add(Dense(512, activation='relu'))
classifier2.add(Dropout(0.2))
classifier2.add(Dense(nb_classes, activation='softmax'))
classifier2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#Eartly stopping to stop for the sufficent number of epochs
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, patience=7, 
        verbose=1, mode='auto', restore_best_weights=False)

classifier2.fit(x_train_seq, y_train, validation_data=(x_val_seq, y_validation_and_test), 
        callbacks=[monitor], epochs=30, batch_size=32, verbose=1)
```

```python
import pickle
pickle.dump(classifier2, open('ANN_trainable_74', 'wb'))
```

# CNN with training the embedding layer

```python
model_cnn_500 = tf.keras.Sequential()
e_cnn_500 = tf.keras.layers.Embedding(vocab_num+1, 100, weights=[embedding_matrix], input_length=max_len, trainable=True)
model_cnn_500.add(e_cnn_500)
model_cnn_500.add(tf.keras.layers.Dropout(0.2))
model_cnn_500.add(tf.keras.layers.Conv1D(filters=100, kernel_size=2, padding='valid', activation='relu', strides=1))
model_cnn_500.add(tf.keras.layers.MaxPooling1D())
model_cnn_500.add(tf.keras.layers.Conv1D(filters=100, kernel_size=2, padding='valid', activation='relu', strides=1))
model_cnn_500.add(tf.keras.layers.GlobalMaxPooling1D())
model_cnn_500.add(tf.keras.layers.Dense(256, activation='relu'))
model_cnn_500.add(tf.keras.layers.Dropout(0.5))
model_cnn_500.add(tf.keras.layers.Dense(nb_classes, activation='softmax'))
model_cnn_500.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#Eartly stopping to stop for the sufficent number of epochs
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, patience=5, 
        verbose=1, mode='auto', restore_best_weights=False)

model_cnn_500.fit(x_train_seq, y_train, validation_data=(x_val_seq, y_validation_and_test), callbacks=[monitor], epochs=30, batch_size=32, verbose=1)
```
