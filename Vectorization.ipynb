# Get the number of words in the vocabulaire

```python
vocab_num = 0
for i in sequences:
    for j in i:
        if vocab_num < j:
            vocab_num = j
            
vocab_num
```

# Embedding index

```python
embeddings_index = {}
for w in model.wv.vocab.keys():
    embeddings_index[w] = model.wv[w]

len(embeddings_index)
```

```python
embeddings_index
```

# Embedding matrix 

```python
embedding_matrix = np.zeros((vocab_num+1, 100))
for word, i in tokenizer.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```
