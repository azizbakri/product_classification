# Class balancing clean (no duplicates)

```python
def clean_balancing(data,min):
    n = data['Class'].count()
    if n < min :
        data = data.iloc[0:0]
    else:
        data = data.head(min)
        
    return data


    
df11 = [ clean_balancing(df_clean[df_clean['Class'] == class_name], 500) for class_name in df_clean.Class.unique()]
df_balanced_no_dupilcates = pd.concat(df11, sort=False, ignore_index=True )
```

# Class balancing

```python
def balancing(data):
    n = data['Class'].count()
    if n < 10 :
        data = pd.concat([data]*10, sort=False, ignore_index=True)
        data = data.head(10)
    elif n < 20 :
        data = pd.concat([data]*20, sort=False, ignore_index=True)
        data = data.head(20)
    elif n < 30 :
        data = pd.concat([data]*30, sort=False, ignore_index=True)
        data = data.head(30)
    elif n < 40 :
        data = pd.concat([data]*40, sort=False, ignore_index=True)
        data = data.head(40)
    elif n < 50 :
        data = pd.concat([data]*50, sort=False, ignore_index=True)
        data = data.head(50)
    elif n < 100 :
        data = pd.concat([data]*100, sort=False, ignore_index=True)
        data = data.head(100)
    elif n >= 100:
        data = data.head(100)
        
    return data


    
df1 = [ balancing(df_clean[df_clean['Class'] == class_name]) for class_name in df_clean.Class.unique()]
df_balanced_100 = pd.concat(df1, sort=False, ignore_index=True )
```
